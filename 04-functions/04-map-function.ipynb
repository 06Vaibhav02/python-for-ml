# Map Function
- The map() function appliles the given function to all items in an input list (or any other)
iterable and returns a map object (an iterator). This is particulary useful for transforming data in a list comprehensively

```python
# syntax map(fn, iterables) - iterables is list, tuples, dictionaries etc

def square(n):
    return n**2

print(square(5))  

#for multiple inputs or list we need to loop, square and append squared item in new list so we use map function directly 
```

```python
## Using Regular function in map()
def square(n):
    return n**2

numbers = [1,2,3,4,5,6]
print(map(square, numbers))

squared_list = list(map(square, numbers))
print(squared_list)
```

```python
## Using Lambda function in map 
numbers = [1,2,3,4,5,6] 
list(map(lambda x: x**2, numbers))
```

```python
## Map multiple iterables
numbers_1 = [1,2,3]
numbers_2 = [4,5,6]

added_numbers = list(map(lambda x,y: x+y, numbers_1, numbers_2))
print(added_numbers)
```

```python
## Map to convert list of strings to integers
str_numbers = ['1','2','3','4','5']
int_numbers = list(map(int, str_numbers))      #int('1') - int('2') - so on
##int is a typecast function - inbuilt function
print(int_numbers) 
```

```python
words = ['apple', 'banana', 'cherry']
upper_words = list(map(str.upper, words))      #words[0].upper() - words[1].upper() - words[2].upper()
print(upper_words)
```

```python
## Map function to list of dictionaries

def get_name(person):           #person is a dictionary
    return person['name']      #for one dictionary

def get_age(person):
    return person['age']

people = [
    {'name':'vaibhav', 'age':28}, 
    {'name':'raj', 'age':27}, 
    {'name':'kaustubh', 'age':26}
]

print(list(map(get_name, people)))
print(list(map(get_age, people)))
#get_name({dict-1}), get_name({dict-2}), get_name({dict-3})  
```
