# Lambda Function
- Lambda funtions are small anonymous functions (no name function) defined using the **lambda** keyword. They can have any number of arguments but only **one expression**. They are commonly used for short operations or as arguments to higher order functions 

```python
# Syntax
lambda arguments: expression
```

```python
def addition(a,b):
    return a+b     ##for only one expression, we defined an entire function 

addition(2,3)
```

```python
addition = lambda a,b: a+b
# print(type(addition))

addition(5,6)
```

```python
def even(number):
    if number%2==0:
        return True
print(even(24))
```

```python
even1 = lambda number: number%2==0
print(even1(54))
print(even1(53))
```

```python
def addition(x,y,z):
    return x+y+z

print(addition(10,20,30))
```

```python
addition1 = lambda x,y,z: x+y+z
print(addition1(5,6,7))
```

```python
## Map Function - applies a function to all items in list

numbers = [1,2,3,4,5,6]

def squared(number):
    return number**2
squared(8)
```

```python
# syntax map(fn, iterables)
numbers = [1,2,3,4,5,6]
# map(lambda x: x**2, numbers) #map at <xxxxxxxxx> - not complete loading need to convert to list
list(map(lambda x: x**2, numbers))
```
