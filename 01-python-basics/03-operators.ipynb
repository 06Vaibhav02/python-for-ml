# Operators and Operations

```python
# ARITHMETIC OPERATORS
a = 2
b = 10

addition = a+b
print("addition = ",addition)

substraction = a-b
print("substraction = ",substraction)

multiplication = a*b
print("multiplication = ",multiplication)

division = b/a
print("divsion = ",division) #quotient #float number

modulo = b%a
print("modulo = ",modulo) #remainder

floorDivision = b//a
print("floor division = ",floorDivision) #quotient #int number

exponent = b**a
print("exponent = ",exponent) 
```

```python
# COMPARISION OPERATORS
a = 10 
b = 10
str1 = "Hello"
str2 = "Hello"

#equal to 
print(a==b)  #true
print(str1==str2) #true

#not equal to 
print(a!=b) #false
print(str1!=str2) #false

c = 10
d = 20

#greater than 
print(c>d) #false

#greater than equal to 
print(c>=d) #false

#less than
print(c<d) #true

#less than equal to
print(c<=d) #true
```

```python
# LOGICAL OPERATORS
x = True
y = True
z = False
e = False

#AND - both true then output true else false
result = x and y
print(result) #true

result = x and z
print(result) #false

#NOT - just negation
result = not x
print(result) #false

result = not z 
print(result) #true

#OR - any one true then output true else false
result = x or y
print(result) #true

result = x or z
print(result) #true

result = e or z
print(result) #false
```
