# Loops

```python
## for loop - iterate over sequence of numbers 
for i in range(5):
    print(i)
```

```python
## for loop with range not starting from 0
for i in range(1,6):
    print(i)
```

```python
for i in range(1,10,2):
    print(i)
```

```python
# for i in range(10,1):
#    print(i)
#above code will not work

for i in range(10,1,-1):
    print(i)
```

```python
for i in range(10,1,-2):
    print(i)
```

```python
# traverse a string using for loop
str = "Vaibhav Patil"
for i in str:
    print(i)
```

```python
# while loop - continues to execute as long as condition is true

count = 0
while count<5:
    print(count)
    count=count+1

```

```python
count = 0
while count%2==0:
    print(count)
    count = count + 1
```

```python
## loop control statements

# break - exits the loop prematurely

for i in range(6):
    if i==3:
        break
    else:
        print(i)
# break from 3  

```

```python
# continue - skips the current iteration only

for i in range(6):
    if i==3:
        continue
    else:
        print(i)
# 3 skipped
```

```python
# pass - it is null operation and does nothing

for i in range(5):
    if i==3:
        pass
    else:
        print(i)
```

```python
# nested loops - a loop inside a loop

for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")        # string interpolation from JS

## inner loop gets completed first next next iteration for outer loop
```

```python
## Examples - calculate the sum first n natural numbers using a while and for loop

number = int(input("Enter a number"))
sum = 0

for i in range(number+1):
    sum = sum + i

print(f"sum of first {number} natural number is {sum}")
```

```python
number = int(input("Enter a number"))
sum = 0
i = 1

while i<=number:
    sum = sum + i
    i = i + 1
print(f"sum of first {number} natural number is {sum}")
```

```python
## Example - display prime numbers between 1 and 100
# prime number - divisible by 1 and iself

for num in range(1,101):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break    
        else:
            print(num)
```
