# Importing, Creating Packages and Modules
- In python, modules and packages help organize and reuse code.

```python
import math          #multiple inbuilt functions inside a package
print(math.sqrt(16))
print(math.sqrt(64))
```

```python
from math import sqrt, pi
print(sqrt(16))            #here no need of math.sqrt() can directly use sqrt
print(sqrt(81)) 

print(pi)
```

```python
import numpy as np     #np - alias, we need not write numpy.xyz() we can directly is np.xyz()
new_array = np.array([1,2,3,4])
print(new_array)

#numpy is not an inbuilt package like math so need to be installed (pip install numpy)
```

```python
from math import *    #import all functions from math
print(sqrt(25))
print(pi)
```
